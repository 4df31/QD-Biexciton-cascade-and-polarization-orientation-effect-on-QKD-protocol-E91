# Analytical model

## Noise(less) model

```python
def correlation_probability(θ:np.float64,ϕ:np.float64):
    return (1./2.)*(np.floa)
def model_analytical(θ:np.float64,β:np.float64,α:np.float64):
    return (1./2.)*(correlation_probability( α ) + correlation_probability( α + β ))
def model_noise_analytical(θ:np.float64,β:np.float64,α:np.float64,Pₑ:np.float64):
    return (1./2.) + (1-Pₑ) * model_analytical(θ,β,α)
```

```python
N_samples = 50
Pₑ=0.001
θ = np.linspace(0,2*π,N_samples)
fig = go.Figure()
fig.add_trace(x = θ, y = model_noise_analytical(θ,0.,0.,Pₑ), name = f"noise percentage = {Pₑ*100}%")
fig.show()
```
