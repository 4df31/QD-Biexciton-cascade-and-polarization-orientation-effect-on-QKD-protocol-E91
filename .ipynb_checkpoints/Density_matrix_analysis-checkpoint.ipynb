```python
import pandas as pd
import numpy as np
from  qutip  import fidelity, Qobj
from scipy.constants import hbar, e
```

```python
"""
=======================================
Fine Structure Energy Measured in μeV
but used  in eV in this notebook
=======================================
"""
#   for 344 QDs
Energy_FSS = 7.92E-6    #   eV
#   for QD A
#Energy_FSS = 2.92E-6   #   eV

"""
==================================
Decay time measured in ps=10⁻¹² s
but used in seconds
==================================
"""
time_XX = 181E-12    #   s
time_X = 255E-12     #   s

total_time = time_XX + time_X

#===================================
hbar_eV = (hbar/e)

#=============================
theta_FSS = total_time * Energy_FSS / hbar_eV 
print(f"relative phase is θ = {theta_FSS/np.pi:1.3f} π = {theta_FSS:1.3f}\n")

#=============================
phase = np.exp(-1j*theta_FSS)
print(f"Re( exp(iθ) ) = {phase.real:1.3f}\nIm( exp(iθ) ) = {phase.imag:1.3f}\n")
```

```python
print(phase)
print(phase.conjugate())
print(phase.conj())
print(np.abs(phase))
```

```python
ideal_phased_state = 0.5*np.array([\
    [1,           0,  0,  phase   ],\
    [0,           0,  0,  0       ],\
    [0,           0,  0,  0       ],\
    [phase.conj(),0,  0,  1       ]],dtype=complex)
ideal_phased_state = Qobj(ideal_phased_state)

for row in ideal_phased_state:
    print('[' + '  '.join(f"{x.real:5.2f}{x.imag:+5.2f}j" for x in row) + ']')
```

```python
density_matrix = pd.read_csv("raw_QST_nature_2024_chen.csv", header=None,sep =',').astype('float64')
re = np.array( density_matrix.iloc[0:4,0:4])
im = np.array(density_matrix.iloc[4:8,0:4])

density_matrix = re + 1j*im
density_matrix = Qobj(density_matrix)


print(f"\nρ = \n")
for row in density_matrix:
    print('[' + '  '.join(f"{x.real:5.2f}{x.imag:+5.2f}j" for x in row) + ']')

"""
print(f"\nTr(ρ) = {np.trace(density_matrix)}")
print(f"\n diag(ρ) = {np.real(np.diag(density_matrix))}")
"""
print(f"{np.abs()den")
```

```python
difference = density_matrix - ideal_phased_state

print(f"\nDifference = \n")
for row in difference.full():
    print('[' + '  '.join(f"{x.real:9.2e}{x.imag:+9.2e}j" for x in row) + ']')
```

```python
density_RR = Qobj(0.25*np.array([\
    [1,-1j,-1j,-1],\
    [1j,1,1,-1j],\
    [1j,1,1,-1j],\
    [-1,1j,1j,1]],dtype=complex))

print(f"\nρ_RR = \n")
for row in density_RR:
    print('[' + '  '.join(f"{x.real:5.2f}{x.imag:+5.2f}j" for x in row) + ']')
density_LL = density_RR.trans().conj()
print(f"\nρ_LL = \n")
for row in density_LL:
    print('[' + '  '.join(f"{x.real:5.2f}{x.imag:+5.2f}j" for x in row) + ']')
density_RL = Qobj(0.25*np.array([\
    [1,1j,-1j,1],\
    [-1j,1,-1,-1j],\
    [1j,-1,1,1j],\
    [1,1j,-1j,1]],dtype=complex))

density_LR = density_RL.trans().conj()
print(f"\nρ_RL = \n")
for row in density_RL:
    print('[' + '  '.join(f"{x.real:5.2f}{x.imag:+5.2f}j" for x in row) + ']')
print(f"\nρ_LR = \n")
for row in density_LR:
    print('[' + '  '.join(f"{x.real:5.2f}{x.imag:+5.2f}j" for x in row) + ']')

noise = 0.25*np.array([density_RR , density_LL , density_RL , density_LR])
```

```python
fidelity_value = fidelity(ideal_phased_state, density_matrix)
print(f"fidelity = {fidelity_value:1.3}")
```

```python
[fidelity(difference,model) for model in noise ]
```
